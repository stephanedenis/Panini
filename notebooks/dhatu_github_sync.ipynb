```python
# ğŸš€ Cellule 2: Chargement Modules GitHub (Auto-MAJ)

def load_latest_modules():
    """Charge les derniÃ¨res versions des modules depuis GitHub"""
    
    print("ğŸ“¥ Chargement modules depuis GitHub...")
    
    # Phase 1: VÃ©rification et prÃ©paration MAJ
    prep_result = hot_reload_manager.check_and_prepare_updates()
    
    if prep_result['updates_available']:
        print(f"ğŸ”„ {len(prep_result.get('modules_to_update', []))} modules mis Ã  jour disponibles")
    else:
        print("âœ… Modules GitHub Ã  jour")
    
    # Import dynamique modules optimaux
    try:
        # DÃ©tection GPU via module GitHub
        from src.modules.gpu.detector import GPUDetector, ModuleSelector
        
        gpu_context = GPUDetector.detect_all_gpu_capabilities()
        analyzer_module = ModuleSelector.get_best_analyzer(gpu_context)
        
        print(f"ğŸ¯ GPU: {gpu_context.get('recommended_mode', 'cpu')}")
        print(f"ğŸ§© Analyseur sÃ©lectionnÃ©: {analyzer_module}")
        
        # Import analyseur adaptatif
        if analyzer_module == 'dhatu_gpu_t4':
            from src.modules.analyzers.dhatu_gpu_t4 import DhatuGPUT4Analyzer
            analyzer = DhatuGPUT4Analyzer()
        else:
            # Fallback analyseur basique
            print("âš ï¸ Utilisation analyseur basique (modules T4 non disponibles)")
            
            # Analyseur minimal intÃ©grÃ©
            class BasicAnalyzer:
                def __init__(self):
                    self.name = "BasicAnalyzer"
                
                def detect_compatibility(self):
                    return {'compatible': True}
                
                def initialize(self, config):
                    return True
                
                def process(self, data):
                    if isinstance(data, str):
                        docs = [data]
                    else:
                        docs = data if isinstance(data, list) else [str(data)]
                    
                    return {
                        'dhatu_analysis': {'total_matches': len(docs) // 3},
                        'molecular_analysis': {'total_molecules': len(docs) // 2},
                        'processing_time': 0.1,
                        'documents_processed': len(docs)
                    }
                
                def get_performance_metrics(self):
                    return {'processing_speed': 50, 'accelerator': 'CPU-Basic'}
                
                def cleanup(self):
                    pass
            
            analyzer = BasicAnalyzer()
        
        # Chargeur de donnÃ©es
        from src.modules.loaders.turbo_gpu_loader import TurboGPULoader
        loader = TurboGPULoader()
        
        # Initialisation
        compatibility = analyzer.detect_compatibility()
        init_success = analyzer.initialize(gpu_context)
        
        print(f"âœ… Analyseur {analyzer.name}: {'ğŸš€' if init_success else 'âš ï¸'}")
        print(f"âœ… Chargeur prÃªt")
        
        return {
            'analyzer': analyzer,
            'loader': loader,
            'gpu_context': gpu_context,
            'github_sync_ready': True
        }
        
    except Exception as e:
        print(f"âŒ Erreur chargement modules: {e}")
        return None

# Chargement initial
modules = load_latest_modules()

if modules:
    analyzer = modules['analyzer']
    loader = modules['loader']
    gpu_context = modules['gpu_context']
    
    print(f"\nğŸ¯ MODULES GITHUB CHARGÃ‰S:")
    print(f"â”œâ”€â”€ Analyseur: {analyzer.name}")
    print(f"â”œâ”€â”€ GPU: {gpu_context.get('recommended_mode', 'cpu')}")
    print(f"â””â”€â”€ Auto-MAJ: Actif")
else:
    print("âŒ Ã‰chec chargement modules GitHub")
```

```python
# âš¡ Cellule 3: Traitement avec Hot-Reload IntÃ©grÃ©

def process_cycle_with_github_sync():
    """Cycle de traitement intÃ©grÃ© avec hot-reload GitHub"""
    
    def core_processing():
        """Fonction de traitement principal"""
        
        print("ğŸ“¥ Chargement donnÃ©es turbo...")
        
        # Configuration adaptÃ©e GPU
        load_config = {
            'max_files': gpu_context.get('optimal_batch_size', 32),
            'max_documents': 200,  # Limite pour dÃ©mo
            'recent_only': True,
            'min_age_minutes': 30,
            'quality_threshold': 0.4
        }
        
        # Chargement donnÃ©es
        data_result = loader.load_data(load_config)
        
        if not loader.validate_data(data_result):
            print("âš ï¸ DonnÃ©es non valides - Utilisation donnÃ©es test")
            # DonnÃ©es de test intÃ©grÃ©es
            test_documents = [
                "Sanskrit dhÄtu analysis with âˆšà¤—à¤®à¥ root",
                "Molecular combination à¤­à¥‚ + à¤…à¤¸à¥à¤¤à¤¿ forms", 
                "Etymological trace from Indo-European âˆšà¤•à¥ƒ",
                "Panini grammar rules for dhÄtu conjugation",
                "Computational linguistics meets ancient Sanskrit"
            ] * 10  # 50 documents test
            
            data_result = {
                'documents': [{'content': doc, 'type': 'test'} for doc in test_documents],
                'metadata': {'test_data': True}
            }
        
        documents = data_result['documents']
        
        print(f"ğŸ“Š {len(documents)} documents Ã  analyser")
        
        # Analyse avec modules GitHub
        start_time = time.time()
        
        # Traitement par petits batches pour dÃ©mo
        batch_size = min(20, len(documents))
        total_processed = 0
        
        for i in range(0, len(documents), batch_size):
            batch = documents[i:i+batch_size]
            
            # Extraction contenu
            batch_content = [doc.get('content', str(doc)) for doc in batch]
            
            # Analyse
            batch_result = analyzer.process(batch_content)
            total_processed += len(batch)
            
            print(f"âš¡ Batch {i//batch_size + 1}: {len(batch)} docs analysÃ©s")
            print(f"   â”œâ”€â”€ DhÄtu: {batch_result['dhatu_analysis']['total_matches']}")
            print(f"   â””â”€â”€ MolÃ©cules: {batch_result['molecular_analysis']['total_molecules']}")
        
        processing_time = time.time() - start_time
        processing_speed = total_processed / processing_time if processing_time > 0 else 0
        
        return {
            'documents_processed': total_processed,
            'processing_time': processing_time,
            'processing_speed': processing_speed,
            'github_modules_used': True
        }
    
    # Workflow intÃ©grÃ© avec hot-reload
    print("ğŸ”„ DÃ©marrage cycle avec GitHub-Sync...")
    
    workflow_result = hot_reload_manager.integrated_cycle_workflow(
        process_function=core_processing,
        cycle_pause_seconds=3  # Pause courte pour dÃ©mo
    )
    
    return workflow_result

# ExÃ©cution du cycle avec GitHub-Sync
if modules:
    cycle_result = process_cycle_with_github_sync()
    
    print(f"\nğŸ† CYCLE GITHUB-SYNC TERMINÃ‰:")
    print(f"â”œâ”€â”€ DurÃ©e totale: {cycle_result.get('total_time', 0):.2f}s")
    print(f"â”œâ”€â”€ SuccÃ¨s: {'âœ…' if cycle_result.get('success') else 'âŒ'}")
    
    if 'phases' in cycle_result:
        phases = cycle_result['phases']
        print(f"â”œâ”€â”€ Phase prÃ©paration: {phases.get('preparation', {}).get('duration', 0):.2f}s")
        print(f"â”œâ”€â”€ Phase traitement: {phases.get('processing', {}).get('duration', 0):.2f}s")
        print(f"â””â”€â”€ Phase hot-reload: {phases.get('hot_reload', {}).get('duration', 0):.2f}s")
    
    # VÃ©rification mises Ã  jour GitHub
    github_status = github_loader.get_module_status()
    print(f"\nğŸ“Š Ã‰tat GitHub:")
    print(f"â”œâ”€â”€ Modules en cache: {github_status.get('cached_modules', 0)}")
    print(f"â””â”€â”€ DerniÃ¨re vÃ©rification: {github_status.get('last_check', 'N/A')[:19]}")
else:
    print("âŒ Impossible d'exÃ©cuter - Modules non chargÃ©s")
```
