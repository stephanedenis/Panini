```python
# ğŸ“¡ Cellule 2: Dashboard Temps RÃ©el - MONITORING CONTINU

def start_realtime_dashboard(refresh_seconds=10, duration_minutes=60):
    """Dashboard temps rÃ©el avec actualisation automatique"""
    
    start_time = time.time()
    duration_seconds = duration_minutes * 60
    
    try:
        while time.time() - start_time < duration_seconds:
            clear_output(wait=True)
            
            # Header avec timestamp
            now = datetime.now()
            elapsed = int(time.time() - start_time)
            remaining = int(duration_seconds - elapsed)
            
            print("ğŸ“¡ DASHBOARD GITHUB-SYNC TEMPS RÃ‰EL")
            print("=" * 60)
            print(f"ğŸ• {now.strftime('%H:%M:%S')} | â±ï¸ Elapsed: {elapsed//60}m{elapsed%60}s | ğŸ”„ Refresh: {refresh_seconds}s | â° Remaining: {remaining//60}m{remaining%60}s")
            print()
            
            # Statut monitoring
            status = hot_reload.get_monitoring_status()
            monitoring_icon = "ğŸŸ¢" if status['monitoring_active'] else "ğŸ”´"
            
            print(f"ğŸ“Š MONITORING STATUS:")
            print(f"â”œâ”€â”€ {monitoring_icon} Monitoring: {'ACTIF' if status['monitoring_active'] else 'ARRÃŠTÃ‰'}")
            print(f"â”œâ”€â”€ ğŸ”„ Auto-reload: {'ON' if status['auto_reload_enabled'] else 'OFF'}")
            print(f"â”œâ”€â”€ ğŸ“¦ Queue: {status['queue_length']} modules")
            print(f"â”œâ”€â”€ âš¡ Reload actif: {'OUI' if status['reload_in_progress'] else 'NON'}")
            print(f"â”œâ”€â”€ ğŸ“ˆ Total reloads: {status['total_reloads']}")
            print(f"â””â”€â”€ ğŸ• Dernier reload: {status['last_reload'] or 'Jamais'}")
            
            # Statut GitHub
            try:
                github_status = loader.get_module_status()
                print(f"\nğŸ™ GITHUB STATUS:")
                print(f"â”œâ”€â”€ ğŸ“¡ Connexion: {'âœ…' if github_status.get('connected') else 'âŒ'}")
                print(f"â”œâ”€â”€ ğŸ“ Modules surveillÃ©s: {len(github_status.get('modules', []))}")
                print(f"â””â”€â”€ ğŸ” DerniÃ¨re vÃ©rif: {github_status.get('last_check', 'Jamais')}")
            except:
                print(f"\nğŸ™ GITHUB STATUS: âŒ Erreur rÃ©cupÃ©ration statut")
            
            # Statistiques reload
            reload_stats = hot_reload.get_detailed_status()
            success_rate = reload_stats.get('success_rate', 0)
            rate_color = "ğŸŸ¢" if success_rate > 80 else "ğŸŸ¡" if success_rate > 50 else "ğŸ”´"
            
            print(f"\nğŸ“ˆ STATISTIQUES RELOAD:")
            print(f"â”œâ”€â”€ {rate_color} Taux succÃ¨s: {success_rate:.1f}%")
            print(f"â”œâ”€â”€ âœ… RÃ©ussis: {reload_stats.get('successful_reloads', 0)}")
            print(f"â”œâ”€â”€ âŒ Ã‰checs: {reload_stats.get('failed_reloads', 0)}")
            print(f"â””â”€â”€ â†©ï¸ Rollbacks: {reload_stats.get('rollbacks', 0)}")
            
            # Actions rapides
            print(f"\nğŸ® ACTIONS RAPIDES:")
            if not status['monitoring_active']:
                print(f"â–¶ï¸ DÃ‰MARRER: hot_reload.start_background_monitoring(5)")
            else:
                print(f"â¹ï¸ ARRÃŠTER: hot_reload.stop_background_monitoring()")
            print(f"ğŸ”„ FORCE RELOAD: hot_reload.manual_reload_trigger()")
            
            # Footer
            print(f"\nğŸ’¡ Dashboard actif - ArrÃªter avec Kernel â†’ Interrupt")
            
            # Attendre avant refresh
            time.sleep(refresh_seconds)
            
    except KeyboardInterrupt:
        clear_output(wait=True)
        print("â¹ï¸ Dashboard arrÃªtÃ© par utilisateur")
        print(f"ğŸ“Š Derniers statuts disponibles via: hot_reload.get_monitoring_status()")

# Lancer dashboard avec paramÃ¨tres par dÃ©faut
print("ğŸš€ LANCEMENT DASHBOARD TEMPS RÃ‰EL")
print("" )
print("ParamÃ¨tres:")
print("- Refresh: 10 secondes")
print("- DurÃ©e: 60 minutes")
print("- ArrÃªt: Kernel â†’ Interrupt")
print("")
print("âš¡ DÃ©marrage dans 3 secondes...")
time.sleep(3)

start_realtime_dashboard(refresh_seconds=10, duration_minutes=60)
```
